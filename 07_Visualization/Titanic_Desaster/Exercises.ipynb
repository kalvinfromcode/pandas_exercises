### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/Visualization/Titanic_Desaster/train.csv). 

### Step 3. Assign it to a variable titanic 

```python
titanic = pd.read_csv('/Users/kalvinc/Projects/pandas-practice/pandas_exercises/07_Visualization/Titanic_Desaster/train.csv', sep=',')
```

### Step 4. Set PassengerId as the index 

```python
titanic.head()
titanic = titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
plt.pie(titanic['Sex'].value_counts(), labels=titanic['Sex'].value_counts().index, autopct='%1.1f%%', 
        explode=(0.15, 0), startangle=90)
plt.axis('equal')
plt.title('Sex Proportion')
plt.tight_layout()
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
titanic.head()
plt.scatter(titanic['Age'], titanic['Fare'], c=titanic['Sex'].map(lambda g : 1 if g == 'male' else 0))

# alt
g = sns.lmplot('Age', 'Fare', data=titanic, hue='Sex', fit_reg=False)
g.axes[0,0].set_ylim(-5)
g.axes[0,0].set_xlim(-5, 85)
```

### Step 7. How many people survived?

```python
titanic[titanic['Survived'] == 1]['Survived'].count()

# alt
titanic.Survived.sum()
```

### Step 8. Create a histogram with the Fare payed

```python
# titanic['Fare'].hist()

# alt
plt.hist(titanic['Fare'].sort_values(ascending=False), range(0, 600, 10))
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Frequency')
plt.show()
```
