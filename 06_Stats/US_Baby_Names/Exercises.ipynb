# US - Baby Names

### Introduction:

We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv). 

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv', sep=',')
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names = baby_names.drop(columns=['Unnamed: 0', 'Id'])
```

### Step 6. Is there more male or female names in the dataset?

```python
'more female' if baby_names['Name'][baby_names['Gender'] == 'F'].nunique() > baby_names['Name'][baby_names['Gender'] == 'M'].nunique() else 'more males'
baby_names['Gender'].value_counts()
```

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby('Name').sum() # .sum() is needed so you can access Count as a standalone column
# remoe Year column
names.head()
```

```python
baby_names.groupby('Name').head()
```

### Step 8. How many different names exist in the dataset?

```python
len(names)
```

### Step 9. What is the name with most occurrences?

```python
# names['Name'].value_counts().sort_values(ascending=False).head(1)
names.Count.idxmax()

# baby_names.groupby('Name').Count.idxmax() is wrong -> will not sort?
```

### Step 10. How many different names have the least occurrences?

```python
len(names[names['Count'] == names.Count.min()]) # idxmin() returns first min
```

### Step 11. What is the median name occurrence?

```python
names[names['Count'] == names.Count.median()]
```

### Step 12. What is the standard deviation of names?

```python
names.std()[1]
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
# del names['Year']
names.describe()
```
